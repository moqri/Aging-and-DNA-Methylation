```python
gse_id='GSE120610'
url='https://ftp.ncbi.nlm.nih.gov/geo/series/GSE120nnn/GSE120610/suppl/GSE120610_Matrix_processed.csv.gz'
!wget {url} -O matrix.gz
```

```python
import pandas as pd
df=pd.read_csv('matrix.gz',index_col=0)
df
```

```python
cols=[c for c in df.columns if 'SAMPLE' in c]
df=df[cols].transpose()
df.index.name = "title"
df
```

```python
import GEOparse
gse = GEOparse.get_GEO(geo=gse_id, destdir="./")
age_field='characteristics_ch1.0.age'
ages=gse.phenotype_data[['description',age_field]]
ages.index=ages.description.str.replace(' ','')
ages=ages.drop('description',1)
ages.columns=['age']
```

```python
df=df.merge(ages,left_index=True, right_index=True)
df.head()
```

```python
import numpy as np
df.age=df.age.replace('-',np.nan)
df=df.dropna(subset=['age'])
df.age=df.age.astype(float)
df.shape
```

```python
cor=df.corrwith(df.age)
cor.sort_values()
```
